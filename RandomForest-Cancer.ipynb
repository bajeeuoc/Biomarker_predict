```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
# Predict probabilities on the test set
y_pred_proba = rf.predict_proba(x_test_scaled)[:, 1]  # Get probabilities for the positive class

# Calculate AUC
auc_value = roc_auc_score(y_test, y_pred_proba)
print("AUC value: ", auc_value)

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {auc_value:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate (1-specificity)', fontsize=14, fontweight='bold')  # Bold X-axis label
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=14, fontweight='bold')    # Bold Y-axis label
plt.title('Random Forest', fontsize=14, fontweight='bold')          # Bold title
plt.legend(loc='lower right', fontsize=12, frameon=True, fancybox=True, shadow=True, prop={'weight':'bold'})  # Bold legend

# Save the plot as pdf
plt.savefig('Random forest_ROC_Curve.pdf')  

# Save the plot as jpg
plt.savefig('Random forest_ROC_Curve.jpg')  # Save as jpg with 300 dpi resolution

plt.show()  # Display the plot
```

```python
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.preprocessing import StandardScaler
import textwrap
```

```python
# Initialize StandardScaler
scaler = StandardScaler()
# Prepare to save AUC results
auc_results = []

# Create a PDF to save all ROC plots
with PdfPages('All_ROC_Curves_RandomForest_Cancer.pdf') as pdf:
    # Iterate over each feature in the dataset
    for column in x.columns:
        # Extract a single feature
        single_feature = x[[column]].values
        
        # Scale the single feature
        single_feature_scaled = scaler.fit_transform(single_feature)
        
        # Train/test split for the current feature
        x_train, x_test, y_train, y_test = train_test_split(single_feature_scaled, y, test_size=0.3, random_state=0, stratify=y)

        # Initialize Random Forest model
        rf_model = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)
        
        # Fit the model with a single feature
        rf_model.fit(x_train, y_train)
        
        # Predict probabilities
        y_pred_proba = rf_model.predict_proba(x_test)[:, 1]
        
        # Calculate AUC
        auc_value = roc_auc_score(y_test, y_pred_proba)
        
        # Store the results
        auc_results.append({'Feature': column, 'AUC': auc_value})
        
        # Calculate ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)

         # Wrap long feature names for the plot title
        wrapped_title = '\n'.join(textwrap.wrap(f'ROC Curve for {column}', width=40))
        
        # Plot ROC curve
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, lw=2, label=f'{column} (AUC = {auc_value:.2f})')
        plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate (1-specificity)')
        plt.ylabel('True Positive Rate (Sensitivity)')
        plt.title(wrapped_title)
        plt.legend(loc="lower right")
        
        # Save the plot to the PDF
        pdf.savefig()
        plt.close()

# Save AUC results to CSV
auc_df = pd.DataFrame(auc_results)
auc_df.to_csv('AUC_Results_Individual_Features_RandomForest_Cancer.csv', index=False)

print("AUC values saved to 'AUC_Results_Individual_Features_RandomForest_Cancer.csv'.")
print("All ROC curves saved to 'All_ROC_Curves_RandomForest_Cancer.pdf'.")
```

```python
from sklearn.metrics import RocCurveDisplay, auc
from sklearn.model_selection import StratifiedKFold
from sklearn.datasets import make_classification
import numpy as np

# Convert X and y to pandas DataFrame for illustration purposes
x = pd.DataFrame(x)
y = pd.Series(y)

# Cross-validation setup
n_splits = 10
cv = StratifiedKFold(n_splits=n_splits)
classifier = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)  # Random Forest

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

fig, ax = plt.subplots(figsize=(6, 6))
# Plot random chance line
ax.plot([0, 1], [0, 1], linestyle='--', color='black', label='Random chance', lw=2)
for fold, (train_idx, test_idx) in enumerate(cv.split(x, y)):
    # Use .iloc to index DataFrames/Series by position
    classifier.fit(x.iloc[train_idx], y.iloc[train_idx])
    viz = RocCurveDisplay.from_estimator(
        classifier,
        x.iloc[test_idx],
        y.iloc[test_idx],
        name=f"ROC fold {fold}",
        alpha=0.3,
        lw=1,
        ax=ax,
    )
    interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
ax.plot(
    mean_fpr,
    mean_tpr,
    color="b",
    label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (mean_auc, std_auc),
    lw=2,
    alpha=0.8,
)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
ax.fill_between(
    mean_fpr,
    tprs_lower,
    tprs_upper,
    color="grey",
    alpha=0.2,
    label=r"$\pm$ 1 std. dev.",
)

ax.set_xlabel("False Positive Rate (1-specificity)", fontsize=14, fontweight='bold')
ax.set_ylabel("True Positive Rate (Sensitivity)", fontsize=14, fontweight='bold')
ax.set_title(f"Cross Validation-Random Forest", fontsize=16, fontweight='bold')

# Adjust legend font size and bold
ax.legend(loc="lower right", fontsize=12, title_fontsize='13', title='Legend', fancybox=True)
# Save the plot as pdf
plt.savefig('Cross Validation-RF_ROC_Curve.pdf')  

# Save the plot as jpg
plt.savefig('Cross Validation-RF_ROC_Curve.jpg')  # Save as jpg with 300 dpi resolution
# Save the plot as png
plt.savefig('Cross Validation-RF_ROC_Curve.png')
plt.show()
```
